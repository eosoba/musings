```python
path = os.getcwd()+'/citydata/state data/'
files=os.listdir(path)
yx = pd.read_csv(path+files[1])
male_profile = list(yx[yx['gender']==2][yx['pref_opento_abstract'].notnull()]['pref_opento_abstract'].values)
female_profile = list(yx[yx['gender']==1][yx['pref_opento_abstract'].notnull()]['pref_opento_abstract'].values)
```

```python
male_profile[:5]
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
```

```python
sifted = [i.lower().translate(None, string.punctuation) for i in male_profile if i != 'NULL']
tokens = []
stopwords = nltk.corpus.stopwords.words('english')
stopwords.append("null")
```

```python
for d in male_profile:
    data_tokens = nltk.wordpunct_tokenize(d)
    filtered_tokens = [w for w in data_tokens if w.lower() not in stopwords]
    tokens = tokens + filtered_tokens
tokens =[i.translate(None,string.punctuation) for i in tokens if len(i.translate(None,string.punctuation)) > 0]
```

```python
tokens[:6]
```

# Find the most significant trigrams in the profiles.

```python
finder = TrigramCollocationFinder.from_words(tokens,5)
finder.apply_freq_filter(5)
best=finder.nbest(trigram_measures.likelihood_ratio, 200)
best
```

# Look at the most significant bigrams

```python
finder = BigramCollocationFinder.from_words(tokens)
finder.apply_freq_filter(10)
best=finder.nbest(bigram_measures.likelihood_ratio, 200)
best
```

```python
# Look the bigrams containing the words 'woman'
xx =['woman','women','girl','lady','female']
[(a,b) for a,b in best if a in xx or b in xx ]
```
